```python
import tensorflow as tf
import pandas as pd
```

```python
data=pd.read_csv('Iris.csv')
data.head()
```

```python
data['y1']=0
data['y2']=0
for i in range(0,len(data['Species'])):
    if(data['Species'][i]=='Iris-versicolor'):
        data['y1'][i]=0
        data['y2'][i]=0
    elif(data['Species'][i]=='Iris-virginica'):
        data['y1'][i]=0
        data['y2'][i]=1
    else:
        data['y1'][i]=1
        data['y2'][i]=0
```

```python
data.drop('Species',axis=1,inplace=True)
```

```python
data.head()
```

```python
x=data.iloc[:,1:5].values
y=data.iloc[:,5:].values
```

```python
y[0]
```

```python
X=tf.placeholder(tf.float32,[None,4])
Y=tf.placeholder(tf.float32,[None,2])
```

```python
w=tf.Variable(tf.zeros([4,2]))
b=tf.Variable(tf.zeros([2]))
```

```python
init=tf.global_variables_initializer()
```

```python
pred=tf.matmul(X,w)+b
pred_s=tf.nn.softmax(pred)
```

```python
cost = tf.reduce_mean(-tf.reduce_sum(Y*tf.log(pred_s), reduction_indices=1))
```

```python
optimizer=tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(1000):
        for xx,yy in zip(x,y):
            loss,_=sess.run([cost,optimizer],feed_dict={X:[xx],Y:[yy]})
        print("Epoch[{}][1000]  Loss: {}".format(epoch,loss))
    weight=sess.run(w)
    bias=sess.run(b)
    #xx=[]
    #prediction=sess.run(pred,feed_dict={X:[xx],w:weight,b:bias})
#     print(prediction)
    #prediction_s=tf.nn.softmax(prediction)
#     print(sess.run(prediction_s))
        
```
